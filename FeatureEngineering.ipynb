```python
import numpy as np

# Selecting a random N between 10 and 15
N = np.random.randint(10, 15)
# 1xN dimensional vector(array)
array = np.random.random(N)

# Calculating mean of sample
mean = sum(array)/N
# Calculating variance of sample
variance = sum((array - mean)**2)/(N-1)

print(f'No. of features(N): {N}')
print(f'Vector array: {array}')
print(f'Mean of the sample: {mean}')
print(f'Variance of the sample: {variance}')
```

```python
import numpy as np

# Selecting a random M between 10 and 15
M = np.random.randint(10, 15)
x = np.random.random(M)
y = np.random.random(M)

# Calculating mean of samples
x_bar = x.mean()
y_bar = y.mean()

# Calculating covarience between x and y
covariance = np.sum((x-x_bar).dot(y-y_bar))/(M-1)

print(f"No. of features(N): {N}")
print(f"Vector X: {x}")
print(f"Vector Y: {y}")
print(f"Mean of Vector X: {x_bar}")
print(f"Mean of Vector Y: {y_bar}")
print(f"Covariance: {covariance}")
```

```python
import numpy as np

# Selecting a random N between 10 and 15
N = np.random.randint(10, 15)

# 1xN dimensional vectors(arrays)
X = np.random.random(N)
Y = np.random.random(N)

# Calculating mean of samples
X_bar = X.mean()
Y_bar = Y.mean()

# Calculating covarience between x and y
covarience = np.sum((X-X_bar).dot(Y-Y_bar)) / (N-1)
var_X = sum((X - X_bar)**2) / (N-1)
var_Y = sum((Y - Y_bar)**2) / (N-1)

# Calculating correlation
correlation = covarience/(np.sqrt(var_X * var_Y))

print(f"No. of features(N): {N}")
print(f"Vector X: {X}")
print(f"Vector Y: {Y}")
print(f"Correlation: {correlation}")
```

```python
import numpy as np

M = 5  # Number of features
N = 10  # Number of samples

# Create a random matrix of shape (M, N)
matrix = np.random.rand(M, N)

# Compute the covariance matrix and correlation matrix
covariance_matrix = np.cov(matrix, rowvar=True, bias=True)
correlation_matrix = np.corrcoef(matrix)

print("Matrix (M x N):")
print(matrix)
print("\nCovariance Matrix:")
print(covariance_matrix)
print("\nCorrelation Matrix:")
print(correlation_matrix)
```
